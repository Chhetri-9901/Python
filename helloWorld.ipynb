```python
arr = [10, 20, 30, 40];
print("This is a demo of list");
print(arr)
arr.append(99);
print(arr);
```

```python
print("This is a demo of Tuples")
arr = (10, 20, 30, 4);
print(arr)
newList = list(arr);
newList.append(88)
print(newList)
finalList = tuple(newList)
print(finalList)
```

```python
print("This is a demo on dictionary")
dic = {
    'firstName': "Ricky",
    'lastName': "Martin",
    'year': 2020,
    'year': 1999
}
print(dic)
print(dic["firstName"])
dic['firstName'] = "Boy"
print(dic)
```
